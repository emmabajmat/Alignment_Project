## Standard Settings
We start with the imports. Some standard imports are required for using the code. If you don't have these, try installing them via pip.

```python
import cv2

from scipy import ndimage
import matplotlib.pyplot as plt

import DataPlotting as DP
import ImageProcessing as IP
```

```python
filename = '../TestImages/TestLines.jpg'
original = cv2.imread(filename, cv2.IMREAD_GRAYSCALE)
```

First, we'll prepare the image for analysis. It's best to align the image horizontally. This makes everything easier later on. Below, you can input values for contrast, the first crop, and the second crop. The first crop isn't always crucial and is mainly for automatically straightening the image. However, sometimes the underlying function is a bit challenging, so it needs some help. The second crop is to trim what you actually want to analyze. Don't be alarmed if a warning pops up.

```python
contrast = 132
crop1 = (190, 175, 200, 150)
crop2 = (500, 375, 200, 200)

bw = cv2.threshold(original,contrast,255,cv2.THRESH_BINARY)[1]
crop = IP.crop_image(bw, *crop1)
rot = ndimage.rotate(original, IP.get_angle(crop))
crop = IP.crop_image(rot, *crop2)
img = cv2.threshold(crop,contrast,255,cv2.THRESH_BINARY)[1]

fig, ax = plt.subplots(1,2, figsize = (15, 5))
ax[0].imshow(original, 'binary')
ax[1].imshow(img, 'gray')
```

Next, you need to determine the scale. This can be done using several built-in functions. Below, they are commented out because we are currently working with a test image.For these examples we will just put it in manually as I know the values! The test images were alle constructed as if they had a SEM scale of 130 microns with size bar of 988 pixels. Deviding these values gives us the microns/pixel.

```python
#scale = get_scale(img)

scale = 130/988
```

## Line width

To show how you can get the line width from these images we wil just use the image above.

```python
height, _ = img.shape

# First we get all the intensity data for the entire image so from row 0 to the max
mean_intensity_profile = IP.get_mean_intensity_profile(img, 0, height)
intensity_profile_derivative = IP.filter_data(IP.get_intensity_profile_dy(mean_intensity_profile))

# We try to get all the edges of the lines
leading_edges = IP.get_leading_edge_positions(intensity_profile_derivative)
trailing_edges = IP.get_trailing_edge_positions(intensity_profile_derivative)
```

```python
DP.draw_mean_and_derivative_plot(
    mean_intensity_profile, 
    intensity_profile_derivative,
    leading_edges,
    trailing_edges
)
```

If we run the cell below you will see the line widths of all the lines in the image and there std. Ofcourse there is no std as they are perfect and they are indeed designed to be 10 microns. A bit lower is the raw data and as you can see technically the computer gives a std but that is to small to actually be of worth.

```python
IP.get_line_width_data(img, mean_intensity_profile, leading_edges, trailing_edges, scale, True)
```
